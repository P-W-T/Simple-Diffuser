# Simple Diffuser

## Inputs

```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from torchvision.datasets import FashionMNIST

from simple_diffuser_improved_cosine_CFG_EMA import *
from diffusion_score import *
```

## Read in datasets

```python
transforms = Compose([
    Resize(32),
    ToTensor(),
    Lambda(lambda x:2*x-1)
])

train_dataset = FashionMNIST("fashion_mnist",  train=True, download=True, transform=transforms)
train_dataloader = DataLoader(train_dataset, batch_size=128, shuffle=True, num_workers=2, pin_memory=True)
```

## Train diffusion model
### Models

```python
#model = UNet(in_channels=1, out_channels=1, time_dim=256, device='cuda')
model = UNet(in_channels=1, out_channels=1, time_dim=256, base_channels=64, n_scaling=3, n_bottleneck=3, device='cuda', num_classes=10)

optimizer = torch.optim.Adam(model.parameters(), lr=3e-4)
diffusion = Diffusion(noise_steps=1000, img_size=32, channels=1, noise_schedule='sigmoid', device='cuda')
```

```python
import numpy as np
for i in train_dataloader:
    plt.imshow(i[0][0][0])
    plt.show()
    plt.imshow(diffusion.noise_images(i[0][0].unsqueeze(0).to('cuda'),torch.tensor([100]))[0].cpu()[0][0])
    plt.show()
    plt.imshow(diffusion.noise_images(i[0][0].unsqueeze(0).to('cuda'),torch.tensor([100]))[1].cpu()[0][0])
    plt.show()
    break
```

### Training

```python
val_samples = train(train_dataloader, model, optimizer, diffusion, epochs=25, device='cuda', target=True, verbose=1, validation=16,
                    img_args={'figsize':(10,10), 'columns':4, 'color':False}, conditional=True, EMA_beta=0.995, EMA_warmup=2000)
```
